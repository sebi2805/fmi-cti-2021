## Exercițiul 1

Fie semnalele continue $x(t) = \cos (520 \pi t + \pi/3)$, $y(t) = \cos (280 \pi t - \pi/3)$ și $z(t) = \cos (120 \pi t + \pi/3)$.

- În Python, simulați axa reală de timp printr-un șir de numere suficient de apropiate, spre exemplu $[0:0.0005:0.03]$ (adică intervalul $[0, 0.03]$, discretizat cu un pas de $0.0005$).

```python
t = np.arange(0, 0.03, 0.0005)

x_t = np.cos(520 * np.pi * t + np.pi / 3)
y_t = np.cos(280 * np.pi * t - np.pi / 3)
z_t = np.cos(120 * np.pi * t + np.pi / 3)

plt.figure(figsize=(12, 8))
```

- Construiți semnalele $x(t)$, $y(t)$ și $z(t)$ și afișați-le grafic, în câte un subplot.

```python
plt.subplot(3, 1, 1)
plt.plot(t, x_t)
plt.title('x(t) ')

plt.subplot(3, 1, 2)
plt.plot(t, y_t)
plt.title('y(t)  ')

plt.subplot(3, 1, 3)
plt.plot(t, z_t)
plt.title('z(t)')


plt.tight_layout()
plt.show()
```

- Eșantionați semnalele cu o frecvență de 200 Hz pentru a obține $x[n]$, $y[n]$ și $z[n]$ și afișați-le grafic, în câte un subplot.

```python
Fs = 200
Ts = 1.0 / Fs

t_200 = np.arange(0, 0.03, Ts)

# Eșantionăm semnalele
x_n = np.cos(520 * np.pi * t_200 + np.pi / 3)
y_n = np.cos(280 * np.pi * t_200 - np.pi / 3)
z_n = np.cos(120 * np.pi * t_200 + np.pi / 3)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.stem(t_200, x_n, 'r', markerfmt='ro', basefmt=" ", linefmt='r')
plt.title('x[n]')

plt.subplot(3, 1, 2)
plt.stem(t_200, y_n, 'g', markerfmt='go', basefmt=" ", linefmt='g')
plt.title('y[n]')

plt.subplot(3, 1, 3)
plt.stem(t_200, z_n, 'b', markerfmt='bo', basefmt=" ", linefmt='b')
plt.title('z[n]')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.tight_layout()
plt.show()
```

## Exercițiul 2

Generați semnale descrise mai jos și afișați-le grafic, fiecare într-un plot. Nu aveți voie să folosiți funcțiile predefinite din `scipy.signal`.

- Un semnal **sinusoidal** de frecvență 400 Hz, care să conțină 1600 de eșantioane.

```python
frecventa = 400
esantioane = 1600

timp = np.linspace(0, 0.03, esantioane, endpoint=False)

semnal = np.sin(2 * np.pi * frecventa * timp)

plt.figure()
plt.xlabel("Timp [s]")
plt.ylabel("Amplitudine")
plt.plot(timp, semnal)
plt.show()
```

- Un semnal **sinusoidal** de frecvență 800 Hz, care să dureze 3 secunde.

```python
frecventa = 800  
timp_total = 3   
esantioane = int(frecventa * timp_total)  

timp = np.linspace(0, timp_total, esantioane, endpoint=False)

# Generarea semnalului sinusoidal
semnal = np.sin(2 * np.pi * frecventa * timp)
plt.figure()
plt.xlabel("Timp [s]")
plt.ylabel("Amplitudine")
plt.plot(timp, semnal)
plt.show()
```

- Un semnal de tip **sawtooth** de frecvență 240 Hz (puteți folosi funcțiile `numpy.floor` sau `numpy.mod`).

```python

frecventa = 240  
timp_total =0.3
esantioane = int(frecventa * timp_total)

timp = np.linspace(0, timp_total, esantioane, endpoint=False)
semnal = np.mod(timp, 1 / frecventa) * frecventa

plt.figure()
plt.title("Semnal Sawtooth")
plt.xlabel("Timp [s]")
plt.ylabel("Amplitudine")
plt.plot(timp, semnal)
plt.show()
```

- Un semnal de tip **square** de frecvență 300 Hz (puteți folosi funcția `numpy.sign`).

```python
import numpy as np
import matplotlib.pyplot as plt

# Definim parametrii
frecventa = 240
esantioane = 1600
T = 1 / frecventa

# Generăm axa timpului
timp = np.linspace(0, 0.03, esantioane, endpoint=False)

# Calculăm semnalul
semnal_sawtooth = 2 * (np.sgn timp / T - np.floor(0.5 + timp / T))

# Afișăm semnalul
plt.figure()
plt.xlabel("Timp [s]")
plt.ylabel("Amplitudine")
plt.plot(timp, semnal_sawtooth)
plt.show()
```

## Exercițiul 3

Un semnal este înregistrat cu o frecvență de eșantionare de 2000 Hz.

- Care este intervalul de timp între două eșantioane?

- Dacă un eșantion este memorat pe 4 biți, câți bytes vor ocupa 1 oră de achiziție?

```python
nr_esantioane=2000
nr_biti=nr_esantioane*4
nr_secunde_per_ora=60*60
#Stiind ca frecventa de esantionare este de 2000 Hz => asta inseamna ca avem 2000 de esantioane pe secunda => 1/2000 => 0.0005 secunde
print(1/nr_esantioane)
#Stiind ca o ora are 60*60 secunde si pe secunda avem 2000 de esantioane => 60*60*2000 esantioane pe ora => un esantion are 4 biti => 60*60*2000*4 biti pe ora
print(nr_biti*nr_secunde_per_ora)
```
