## Exercițiul 1

Realizați graficele din figurile 1 și 2 din îndrumar pentru un semnal sinusoidal cu o frecvență aleasă de voi, _alta_ decât cea utilizată aici.

Reamintim că graficul din dreapta din figura 1 reprezintă înfășurarea semnalului pe cercul unitate, anume reprezentarea în planul complex a șirului $y[n] = x[n] \cdot e^{-2 \pi i n}$.

De asemenea, figura 2 arată influența diferitelor frecvențe de înfășurare asupra formei pe care o are această reprezentare. \\
Afișați grafic $z[\omega] = x[n] \cdot e^{-2 \pi i \omega n}$, pentru **patru valori diferite** ale $\omega$, dintre care una egală cu frecvența semnalului.

```python
fs = 100   
f = 10    

t = np.arange(0, 1, 1/fs)
x = np.sin(2 * np.pi * f * t)

n = np.arange(len(x))
y = x * np.exp(-2j * np.pi * n / len(x))

plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(t, x, label='Sinusoidal Signal')
plt.scatter(t[62], x[62], color='red') 
plt.title('Sinusoidal Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(np.real(y), np.imag(y), label='Wrapped Signal')
plt.scatter(np.real(y[62]), np.imag(y[62]), color='red') 
plt.title('Signal Wrapping on the Unit Circle')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.grid(True)
plt.legend()
plt.axis('equal')

plt.tight_layout()
plt.show()
```

```python
omega_values = [3, 5, 7, 9]
f=5
n_fine = np.arange(0, 1, 0.001)   

x_n_fine = np.sin(2 * np.pi * f * n_fine)

plt.figure(figsize=(12, 8))

for i, omega in enumerate(omega_values, 1):
    z_omega_n_fine = x_n_fine * np.exp(-2j * np.pi * omega * n_fine)

    plt.subplot(2, 2, i)
    plt.plot(np.real(z_omega_n_fine), np.imag(z_omega_n_fine))
    plt.title(f'Reprezentare pentru ω = {omega} Hz')
    plt.xlabel('Partea Reală')
    plt.ylabel('Partea Imaginară')
    plt.axis('equal')

plt.tight_layout()
plt.show()
```
