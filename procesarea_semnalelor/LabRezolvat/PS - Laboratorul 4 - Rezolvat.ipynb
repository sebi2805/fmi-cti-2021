## Exercițiul 1

Prespuneți că într-o incintă temperatura este controlată astfel încât să se obțină o periodicitate noapte-zi asemănătoare unei sinusoide; maximul de temperatură ($1^{\circ} \text{C}$) se atinge la ora 12:00, minimul ($-1^{\circ} \text{C}$) la ora 24:00, nu există variații între zile. Începeți să măsurați temperatura din incintă într-o zi de miercuri, la ora 12:00.

a) Dacă măsurați temperatura la fiecare 2 ore, care va fi **frecvența de eșantionare** (exprimată **în Hz**)?

```python
perioada_esantionare_ore = 2
perioada_esantionare_secunde = perioada_esantionare_ore * 3600
frecventa_esantionare_herzi = 1 / perioada_esantionare_secunde

print("Frecvența de eșantionare va fi", frecventa_esantionare_herzi, "Hz")
```

b) Care este **perioada** semnalului de temperatură și care este **frecvența** acestuia? Este eșantionarea din subpunctul precedent **corectă** în sensul teoremei Nyquist? Care este intervalul de perioada maximă la care trebuie măsurată temperatura astfel încât să îndeplinească criteriul de eșantionare Nyquist?

```python
perioada_semnal_ore = 24
perioada_semnal_secunde = perioada_semnal_ore * 3600
frecventa_semnal = 1 / perioada_semnal_secunde

frecventa_nyquist = 2 * frecventa_semnal

if frecventa_esantionare_herzi > frecventa_nyquist:
    print("Eșantionarea este corectă conform Nyquist,", frecventa_esantionare_herzi, "Hz > 2 *", frecventa_semnal, "Hz")

perioada_maxima = 1 / frecventa_nyquist

print("Perioada maximă de eșantionare este", perioada_maxima, "secunde, adică", perioada_maxima / 3600, "ore")
```

## Exercițiul 3

Calculați frecvența optimă (cea mai eficientă) de eșantionare sub-Nyquist pentru un semnal de bandă $B = 10 \, \text{Hz}$ centrat în jurul lui $f_c = 90 \, \text{Hz}$ pentru următoarele 3 valori posibile ale numărului de replici: a) m = 1, b) m = 2, c) m = 4.

```python
B = 10
f_c = 90

for m in (1, 2, 4):
    maxim = (2 * f_c - B) / m
    minim = (2 * f_c + B) / (m + 1)

    # Trebuie să fie și > 2 * B
    minim = max(minim, 2 * B)

    print("Intervalul de frecvențe posibile:", f"[{minim} Hz, {maxim} Hz]")
    print("Frecvența optimă din interval:", minim, "Hz")
```

## Exercițiul 4

Puterea unui semnal este $P_{semnal} = 90 \, \text{dB}$. Se cunoaște raportul semnal-zgomot, $\text{SNR}_{\text{dB}} = 80 \, \text{dB}$. Care este puterea zgomotului?

**Indicație:** deoarece nu știți puterea de referință față de care se măsoară aceste puteri în decibeli, ar trebui să găsiți răspunsul fără a face conversia în watts.

Din definiție, avem că
$$
\begin{align*}
    \text{SNR}_{\text{dB}} &= 10 \log_{10} (\text{SNR}) = 10 \log_{10} \left(\frac{P_{semnal}}{P_{zgomot}}\right) \\[1em]
    &= 10 \left(\log_{10} \left(P_{semnal}\right) - \log_{10} \left(P_{zgomot}\right)\right) \\[1em]
    &= 10 \left(\left(\log_{10} \left(P_{semnal}\right) - \log_{10} \left(P_0\right)\right) - \left(\log_{10} \left(P_{zgomot}\right) - \log_{10} \left(P_0\right)\right)\right) \\[1em]
    &= 10 \log_{10} \left(\frac{P_{semnal}}{P_0}\right) - 10 \log_{10} \left(\frac{P_{zgomot}}{P_0}\right) \\[1em]
    &= P_{semnal, \,\text{dB}} - P_{zgomot, \, \text{dB}}
\end{align*}
$$

Deci $P_{zgomot} = 90 \, \text{dB} - 80 \, \text{dB} = 10 \, \text{dB}$.
