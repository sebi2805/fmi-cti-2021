## Exercițiul 1

Fie semnalele continue $x(t) = \cos (520 \pi t + \pi/3)$, $y(t) = \cos (280 \pi t - \pi/3)$ și $z(t) = \cos (120 \pi t + \pi/3)$.

- În Python, simulați axa reală de timp printr-un șir de numere suficient de apropiate, spre exemplu $[0:0.0005:0.03]$ (adică intervalul $[0, 0.03]$, discretizat cu un pas de $0.0005$).

```python
timp = np.arange(0, 0.03, 0.0005)
timp
```

- Construiți semnalele $x(t)$, $y(t)$ și $z(t)$ și afișați-le grafic, în câte un subplot.

```python
def x(t):
    return np.cos(520 * np.pi * t * np.pi / 3)

def y(t):
    return np.cos(280 * np.pi * t - np.pi / 3)

def z(t):
    return np.cos(120 * np.pi * t + np.pi / 3)

esantioane_x = x(timp)
esantioane_y = y(timp)
esantioane_z = z(timp)

fig, axes = plt.subplots(3, 1)
axes[0].plot(timp, esantioane_x)
axes[1].plot(timp, esantioane_y)
axes[2].plot(timp, esantioane_z)
plt.show()
```

- Eșantionați semnalele cu o frecvență de 200 Hz pentru a obține $x[n]$, $y[n]$ și $z[n]$ și afișați-le grafic, în câte un subplot.

```python
timp_nou = np.arange(0, 0.03, 1/200)

esantioane_x = x(timp_nou)
esantioane_y = y(timp_nou)
esantioane_z = z(timp_nou)

fig, axes = plt.subplots(3, 1)
axes[0].plot(timp_nou, esantioane_x)
axes[1].plot(timp_nou, esantioane_y)
axes[2].plot(timp_nou, esantioane_z)
plt.show()
```

## Exercițiul 2

Generați semnale descrise mai jos și afișați-le grafic, fiecare într-un plot. Nu aveți voie să folosiți funcțiile predefinite din `scipy.signal`.

- Un semnal **sinusoidal** de frecvență 400 Hz, care să conțină 1600 de eșantioane.

```python
timp = np.linspace(0, 0.01, 1600)
semnal = np.sin(2 * np.pi * 400 * timp)

plt.plot(timp, semnal)
plt.show()
```

- Un semnal **sinusoidal** de frecvență 800 Hz, care să dureze 3 secunde.

```python
timp = np.linspace(0, 3, 2001)
semnal = np.sin(2 * np.pi * 800 * timp)

plt.plot(timp, semnal)
plt.show()
```

- Un semnal de tip **sawtooth** de frecvență 240 Hz (puteți folosi funcțiile `numpy.floor` sau `numpy.mod`).

```python
timp = np.linspace(0, 0.05, 2000)
semnal = np.mod(2 * np.pi * 240 * timp, 2 * np.pi)

plt.plot(timp, semnal)
plt.show()
```

- Un semnal de tip **square** de frecvență 300 Hz (puteți folosi funcția `numpy.sign`).

```python
timp = np.linspace(0, 0.05, 1000)
semnal = np.sign(np.sin(2 * np.pi * 300 * timp))

plt.plot(timp, semnal)
plt.show()
```

## Exercițiul 3

Un semnal este înregistrat cu o frecvență de eșantionare de 2000 Hz.

- Care este intervalul de timp între două eșantioane?

- Dacă un eșantion este memorat pe 4 biți, câți bytes vor ocupa 1 oră de achiziție?

```python
frecventa_de_esantionare = 2000 # Hz

print("Intervalul de timp dintre două eșantioane este", 1/frecventa_de_esantionare, "secunde")

durata_ore = 1
durata_secunde = durata_ore * 3600

numar_esantioane = durata_secunde * frecventa_de_esantionare

dimensiune_biti = 4 * numar_esantioane
dimensiune_bytes = dimensiune_biti / 8

print("O oră de achiziție ocupă", dimensiune_bytes, "bytes")
```
