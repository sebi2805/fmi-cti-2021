## Exercițiul 1

Realizați graficele din figurile 1 și 2 din îndrumar pentru un semnal sinusoidal cu o frecvență aleasă de voi, _alta_ decât cea utilizată aici.

Reamintim că graficul din dreapta din figura 1 reprezintă înfășurarea semnalului pe cercul unitate, anume reprezentarea în planul complex a șirului $y[n] = x[n] \cdot e^{-2 \pi i n}$.

De asemenea, figura 2 arată influența diferitelor frecvențe de înfășurare asupra formei pe care o are această reprezentare. \
Afișați grafic $z[\omega] = x[n] \cdot e^{-2 \pi i \omega n}$, pentru **patru valori diferite** ale $\omega$, dintre care una egală cu frecvența semnalului.

```python
timp = np.linspace(0, 1, 300)

frecventa_semnal = 5
semnal = np.sin(2 * np.pi * frecventa_semnal * timp)

def infasurare(x, omega):
    n = np.linspace(0, 1, len(x))
    return x * np.exp(- 2 * np.pi * 1j * omega * n)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].plot(timp, semnal)

puncte = infasurare(semnal, 1)
axes[1].plot(puncte.real, puncte.imag)

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

puncte = infasurare(semnal, 2)
axes[0][0].plot(puncte.real, puncte.imag)

puncte = infasurare(semnal, 3)
axes[0][1].plot(puncte.real, puncte.imag)

puncte = infasurare(semnal, 5)
axes[1][0].plot(puncte.real, puncte.imag)

puncte = infasurare(semnal, 12)
axes[1][1].plot(puncte.real, puncte.imag)

plt.show()
```

## Exercițiul 2

Afișați **modulul** (valoarea absolută) a transformatei Fourier (folosind relația 1 din PDF) pentru un semnal compus de voi, având **cel puțin trei componente de frecvență distincte** (obțineți un grafic asemănător figurii 3).

Ajustați frecvențele de înfășurare $\omega$ utilizate în transformata Fourier în funcție de frecvența caracteristică a sinusoidei.

```python
timp = np.linspace(0, 0.3, 1000)

amplitudine1 = 2
frecventa1 = 10
faza1 = 0

amplitudine2 = 0.8
frecventa2 = 70
faza2 = 3 * np.pi / 2

amplitudine3 = 1.2
frecventa3 = 30
faza3 = np.pi / 6

semnal = (
    amplitudine1 * np.sin(2 * np.pi * frecventa1 * timp + faza1) +
    amplitudine2 * np.sin(2 * np.pi * frecventa2 * timp + faza2) +
    amplitudine3 * np.sin(2 * np.pi * frecventa3 * timp + faza3)
)

plt.plot(timp, semnal)
plt.show()
```

```python
frecventa_maxima = 30

transformata = []
for omega in range(1, frecventa_maxima + 1):
    modul = np.abs(np.sum(infasurare(semnal, omega)))
    transformata.append(modul)

plt.plot(transformata)
plt.show()
```
